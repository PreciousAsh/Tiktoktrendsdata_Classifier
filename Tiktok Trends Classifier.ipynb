```python
def socialmedia (Platform):
    if len(Platform)== 1:
        Platform.append('None')
    return {'first_list': Platform[0], 'second_list' : Platform[1]}
```

```python
#Zip combinnes 2 coloums
#Tuples are pairs
#Syntax for tuples is (list1, list2)
#([i win],NaN)
features_and_labels = zip(tiktok.Trend,tiktok.Music)
```

```python
#feature_sets = [ x for x in list]
feature_sets = [(socialmedia(Trend), Music) for (Trend, Music) in features_and_labels]
feature_sets
```

```python
import random
random.shuffle(feature_sets)
```

```python
len(feature_sets)
```

```python
training_set = feature_sets[:21]
testing_set = feature_sets[21:]
```

```python
import nltk
```

```python
classifier = nltk.NaiveBayesClassifier.train(training_set)
```

```python
classifier.classify(socialmedia(['Tiktok', 'Youtube_Shorts']))
```

```python
print(nltk.classify.accuracy(classifier, testing_set))
```
